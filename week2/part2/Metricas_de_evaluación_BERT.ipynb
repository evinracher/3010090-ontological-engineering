##BLEU (Bilingual Evaluation Understudy)
Es la métrica "abuela" del NLP. Su enfoque principal es la precisión.

- ¿Cómo funciona? Cuenta cuántas palabras (o grupos de palabras llamados n-gramas) de la predicción aparecen también en la referencia.

- El punto clave: Si el modelo escribe palabras que no están en la referencia, el puntaje baja.

- Limitación: Es muy rígida. Si el humano escribió "feliz" y la IA escribió "contento", BLEU dirá que es un error porque no son la misma palabra exacta.

##ROUGE (Recall-Oriented Understudy for Gisting Evaluation)
A diferencia de BLEU, ROUGE se enfoca en el Recall (recuerdo). Es el estándar de oro para evaluar resúmenes.

- ¿Cómo funciona? Mide cuántas palabras de la referencia humana fueron capaces de aparecer en la predicción de la IA.

- Variantes comunes:

  - ROUGE-1: Coincidencia de palabras sueltas.

  - ROUGE-2: Coincidencia de pares de palabras (bigramas).

  - ROUGE-L: Busca la secuencia más larga de palabras que coincidan en orden (Longest Common Subsequence).

- Utilidad: Es genial para saber si la IA "olvidó" información importante que estaba en el texto original.

##BERTScore
Esta es la métrica moderna y "pesada". A diferencia de las anteriores, no mira letras, sino significados.

- ¿Cómo funciona? Utiliza un modelo BERT para convertir cada palabra en un embedding (un vector numérico que representa su significado). Luego, calcula la similitud entre esos vectores.

- La gran ventaja: Entiende la semántica. Si la predicción usa sinónimos o cambia el orden de la frase pero mantiene el sentido, BERTScore le dará una puntuación alta.

- Comparación: Mientras BLEU y ROUGE son como un profesor que corrige ortografía con una regla, BERTScore es como un profesor que lee el ensayo y evalúa si entendiste el concepto.

```python
import evaluate

# 1. Cargamos las métricas desde la librería 'evaluate' de Hugging Face
bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")
bertscore = evaluate.load("bertscore")

# 2. Definimos los textos a comparar
# Imagina que evaluamos un modelo que traduce del inglés al español
referencias = ["El gato está sentado sobre la alfombra roja"]
predicciones = ["El gato se encuentra sentado en la alfombra de color rojo"]

# --- CÁLCULO DE MÉTRICAS ---

# BLEU: Mide la precisión (qué tanto de la predicción está en la referencia)
results_bleu = bleu.compute(predictions=predicciones, references=referencias)

# ROUGE: Mide el recall (qué tanto de la referencia logró capturar la predicción)
results_rouge = rouge.compute(predictions=predicciones, references=referencias)

# BERTSCORE: Mide la similitud semántica usando embeddings de BERT
results_bertscore = bertscore.compute(
    predictions=predicciones,
    references=referencias,
    lang="es" # Importante especificar el idioma
)

# --- RESULTADOS ---

print(f"--- Resultados de Evaluación ---")
print(f"BLEU Score: {results_bleu['bleu']:.4f}")
print(f"ROUGE-L: {results_rouge['rougeL']:.4f}")
print(f"BERTScore (F1): {results_bertscore['f1'][0]:.4f}")
```
