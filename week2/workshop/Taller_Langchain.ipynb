## 2) Sistemas Workflow

Un workflow es una secuencia de pasos predefinidos (ms control).
Ejemplo: **Clasificar intencin  Ejecutar accin**.

Vamos a hacer un workflow: *clasificar* si el usuario quiere **definicin** o **ejemplo**.


```python

from pydantic import BaseModel, Field
from typing import Literal
from langchain_core.runnables import RunnableBranch, RunnableLambda

class Intent(BaseModel):
    intent: Literal["definicion", "ejemplo", "resume"] = Field(description="Tipo de respuesta deseada")
    reason: str = Field(description="Razn breve")

router_prompt = ChatPromptTemplate.from_messages([
    ("system",
     "Clasifica la intencin. Responde SOLO con JSON del esquema."
     "Usa 'definicion' si pide explicacin, 'ejemplo' si pide cdigo o ejemplo, y 'resume' si pide un resumen"),
    ("user", "{question}")
])

router = router_prompt | llm.with_structured_output(Intent)

definicion_chain = ChatPromptTemplate.from_messages([
    ("system", "Explica en 4 lneas."),
    ("user", "{question}")
]) | llm | StrOutputParser()

ejemplo_chain = ChatPromptTemplate.from_messages([
    ("system", "Da un ejemplo corto en Python."),
    ("user", "{question}")
]) | llm | StrOutputParser()

resume_chain = ChatPromptTemplate.from_messages([
    ("system", "Da un resumen en 3 lineas del tema ingresado"),
    ("user", "{question}")
]) | llm | StrOutputParser()

workflow = (
    RunnableLambda(lambda x: {"question": x["question"], "route": router.invoke({"question": x["question"]})})
    | RunnableLambda(lambda x: {"question": x["question"], "intent": x["route"].intent})
    | RunnableBranch(
        (lambda x: x["intent"] == "ejemplo", ejemplo_chain),
        (lambda x: x["intent"] == "resume", resume_chain),
        definicion_chain
    )
)

print(workflow.invoke({"question": "Explcame qu es una cadena (chain) en LangChain"}))
print("\n---\n")
print(workflow.invoke({"question": "Dame un ejemplo en Python de una chain sencilla"}))
print("\n---\n")
print(workflow.invoke({"question": "Dame un resumen de como funciona una PH"}))
```

## 4) ChatModels

Un Chat Model es una abstraccin de LangChain que representa un modelo de lenguaje optimizado para conversaciones. A diferencia de los modelos de completado de texto, los chat models entienden el contexto de una conversacin con mensajes de diferentes roles (usuario, asistente, sistema).

### Tipos de Mensajes

LangChain define varios tipos de mensajes:

- **SystemMessage**: Instrucciones iniciales que guan el comportamiento del modelo
- **HumanMessage**: Mensajes del usuario
- **AIMessage**: Respuestas del modelo
- **FunctionMessage** / **ToolMessage**: Mensajes relacionados con llamadas a funciones
- **ChatMessage**: Mensaje genrico con rol personalizado

### Uso de SystemMessage

Los mensajes de sistema son fundamentales para definir el comportamiento y personalidad del modelo. Veamos cmo influyen en las respuestas.

```python
# Ejemplo con mensaje de sistema
mensajes = [
    SystemMessage(content="""Eres un meteorlogo experto especializado en anlisis climtico.
    Responde de forma tcnica pero comprensible, siempre incluyendo datos cientficos."""),
    HumanMessage(content="Por qu llueve ms en algunas regiones que en otras?")
]

respuesta = llm.invoke(mensajes)
print("Respuesta del meteorlogo experto:")
print(respuesta.content)
print("\n" + "="*50 + "\n")

# Ahora sin mensaje de sistema
mensajes_simple = [
    HumanMessage(content="Por qu llueve ms en algunas regiones que en otras?")
]

respuesta_simple = llm.invoke(mensajes_simple)
print("Respuesta sin contexto de sistema:")
print(respuesta_simple.content)
```

### Cadenas Secuenciales

Podemos crear cadenas ms complejas donde la salida de una se convierte en la entrada de la siguiente.

```python
from langchain_core.runnables import RunnablePassthrough

# Primera cadena: genera un pronstico meteorolgico
prompt_pronostico = ChatPromptTemplate.from_messages([
    ("system", "Eres un meteorlogo. Genera un pronstico tcnico breve para la ciudad solicitada."),
    ("human", "Ciudad: {ciudad}")
])

cadena_pronostico = prompt_pronostico | llm | StrOutputParser()

# Segunda cadena: traduce el pronstico a lenguaje simple
prompt_simplificacion = ChatPromptTemplate.from_messages([
    ("system", "Eres un comunicador experto. Toma informacin tcnica y explcala de forma que cualquier persona pueda entenderla."),
    ("human", "Informacin tcnica: {pronostico_tecnico}")
])

cadena_simplificacion = prompt_simplificacion | llm | StrOutputParser()

# Combinar ambas cadenas
# RunnablePassthrough permite pasar datos entre cadenas
cadena_completa = (
    {"pronostico_tecnico": cadena_pronostico}
    | cadena_simplificacion
)

# Ejecutar la cadena completa
resultado_final = cadena_completa.invoke({"ciudad": "Bogot"})

print("Pronstico simplificado:")
print(resultado_final)
```

## 7) Memoria

En LangChain una forma comn es usar `RunnableWithMessageHistory`.
Esto permite recordar mensajes previos entre llamadas (mismo `session_id`).


```python
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser

store = {}  # session_id -> history

def get_history(session_id: str):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]

memory_prompt = ChatPromptTemplate.from_messages([
    ("system", "Eres un asistente. Usa el historial de conversacin para mantener contexto."),
    MessagesPlaceholder("history"),
    ("human", "{input}"),
])

memory_chain = memory_prompt | llm | StrOutputParser()

chat_with_memory = RunnableWithMessageHistory(
    memory_chain,
    get_session_history=get_history,
    input_messages_key="input",
    history_messages_key="history",
)

session_id = "kevin-demo-1"

print(chat_with_memory.invoke(
    {"input": "Mi nombre es Kevin. Recurdalo."},
    config={"configurable": {"session_id": session_id}}
))

print(chat_with_memory.invoke(
    {"input": "Cmo me llamo?"},
    config={"configurable": {"session_id": session_id}}
))
```

## 8) Herramientas (Tools)

Las tools permiten que el agente ejecute acciones (clculo, BD, APIs).
Aqu haremos una tool de calculadora y una de hora, y un agente que decide usarlas.


```python
from langchain_core.tools import tool
from langchain.agents import create_agent
import re


@tool
def contar_palabras(texto: str) -> str:
    """Cuenta cuntas palabras tiene el texto."""
    palabras = texto.split()
    return f"TOTAL_PALABRAS: {len(palabras)}"


SYSTEM_PROMPT = """
Eres un agente ReAct (Reason + Act) para demos de LangChain.

Reglas:
- Si el usuario pide contar palabras o analizar un texto,
  usa la tool contar_palabras.
- Si no hace falta herramienta, responde directo.
- Respuesta final: breve, clara y con resultado.
- No expliques tus reglas internas.
"""

agent = create_agent(
    model=llm,
    tools=[contar_palabras],
    system_prompt=SYSTEM_PROMPT,
)

result = agent.invoke({
    "messages": [{
        "role": "user",
        "content": "Cuenta cuntas palabras tiene este texto: LangChain permite crear agentes con herramientas."
    }]
})

print(result["messages"][-1].content)

```

## 10) Streaming

```python

import sys
from langchain_core.prompts import ChatPromptTemplate

stream_prompt = ChatPromptTemplate.from_messages([
    ("system", "Responde en espaol y de forma breve."),
    ("user", "Resume el siguiente tema: {topic}")
])

stream_chain = stream_prompt | llm  # sin parser para streaming real

print(" Streaming:\n")
for chunk in stream_chain.stream({"topic": "qu es LangChain y para qu se usa"}):
    if getattr(chunk, "content", None):
        sys.stdout.write(chunk.content)
        sys.stdout.flush()
print()
```

## 11) Routing

Routing elige **qu pipeline ejecutar** segn la intencin.
Usaremos salida estructurada y `RunnableBranch`.


```python
from pydantic import BaseModel, Field
from typing import Literal
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableBranch
from langchain_core.output_parsers import StrOutputParser


class RouteDecision(BaseModel):
    route: Literal["teoria", "ejercicios"] = Field(
        description="Tipo de respuesta acadmica requerida"
    )
    reason: str = Field(description="Razn breve de la clasificacin")


router_prompt = ChatPromptTemplate.from_messages([
    ("system",
     "Eres un clasificador acadmico.\n"
     "Decide la intencin del estudiante.\n"
     "- Usa 'teoria' si pide definicin, concepto o explicacin.\n"
     "- Usa 'ejercicios' si pide prctica, problemas o actividades.\n"
     "Responde SOLO con JSON vlido."
    ),
    ("user", "Pregunta del estudiante: {question}")
])

router = router_prompt | llm.with_structured_output(RouteDecision)


# Cadena para teora
teoria_chain = ChatPromptTemplate.from_messages([
    ("system", "Responde como profesor universitario. Explica claro y estructurado."),
    ("user", "{question}")
]) | llm | StrOutputParser()


# Cadena para ejercicios
ejercicios_chain = ChatPromptTemplate.from_messages([
    ("system",
     "Genera 3 ejercicios prcticos relacionados con la pregunta.\n"
     "Incluye dificultad (bsico/intermedio/avanzado)."
    ),
    ("user", "{question}")
]) | llm | StrOutputParser()


# Pipeline de routing
routing_pipeline = (
    (lambda x: {
        "question": x["question"],
        "decision": router.invoke({"question": x["question"]})
    })
    | RunnableBranch(
        (lambda x: x["decision"].route == "ejercicios", ejercicios_chain),
        teoria_chain
    )
)


# Pruebas
print(routing_pipeline.invoke({
    "question": "Explcame qu es memoria en LangChain."
}))

print("\n---\n")

print(routing_pipeline.invoke({
    "question": "Dame ejercicios para practicar routing en LangChain."
}))
```
