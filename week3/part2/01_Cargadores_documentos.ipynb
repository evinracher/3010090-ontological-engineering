## Carga de documentos Word (DOCX)

En este ejemplo crearemos un archivo **.docx** de muestra en Colab, lo cargaremos y luego podremos usar Gemini para resumirlo o extraer puntos clave.


```python
# (Ejemplo) Crear un DOCX de muestra en el entorno
from docx import Document

ruta_docx = "ejemplo_documento.docx"

doc = Document()
doc.add_heading("Tecnologías emergentes (ejemplo)", level=1)
doc.add_paragraph("Este documento es un ejemplo para demostrar cómo cargar archivos Word (DOCX).")
doc.add_paragraph("Incluye ideas sobre IA generativa, RAG, agentes y gobernanza de datos.")
doc.add_paragraph("Caso de uso: En una empresa, RAG permite responder con información actualizada desde documentos internos.")

doc.save(ruta_docx)

print("✅ DOCX creado:", ruta_docx)
```

```python
# Cargar DOCX (opción 1) con LangChain: Docx2txtLoader
from langchain_community.document_loaders import Docx2txtLoader

loader_docx = Docx2txtLoader(ruta_docx)
docs_docx = loader_docx.load()

print("Documentos cargados:", len(docs_docx))
print("\n--- Vista previa ---\n")
print(docs_docx[0].page_content[:600])
```

## Carga de documentos Excel (XLSX)

Para Excel, una forma ligera y muy común es **leer con pandas** y convertir el contenido a texto (por ejemplo, CSV/Markdown) para enviarlo al LLM.


```python
# (Ejemplo) Crear un XLSX de muestra
import pandas as pd

ruta_xlsx = "ejemplo_ventas.xlsx"

df = pd.DataFrame({
    "mes": ["Enero", "Febrero", "Marzo", "Abril"],
    "producto": ["A", "A", "B", "B"],
    "ventas": [120, 150, 90, 110],
    "ciudad": ["Medellín", "Bogotá", "Cali", "Medellín"]
})

df.to_excel(ruta_xlsx, index=False)
print("✅ XLSX creado:", ruta_xlsx)
```

```python
# Cargar XLSX con pandas y convertir a texto
df_cargado = pd.read_excel(ruta_xlsx)

# Convertimos la tabla a un texto que el LLM pueda usar como contexto
texto_excel = df_cargado.to_markdown(index=False)

print(texto_excel)
```

## Carga de archivos JSON

LangChain incluye cargadores para JSON. Aquí creamos un JSON simple, lo cargamos y revisamos su contenido.


```python
# Instalar dependencia requerida
!pip -q install jq
```

```python
# (Ejemplo) Crear un JSON de muestra
import json

ruta_json = "ejemplo_clientes.json"

data_json = {
    "clientes": [
        {"id": 1, "nombre": "Ana", "segmento": "Premium", "ciudad": "Medellín"},
        {"id": 2, "nombre": "Luis", "segmento": "Estándar", "ciudad": "Bogotá"},
        {"id": 3, "nombre": "Sofía", "segmento": "Premium", "ciudad": "Cali"}
    ]
}

with open(ruta_json, "w", encoding="utf-8") as f:
    json.dump(data_json, f, ensure_ascii=False, indent=2)

print("✅ JSON creado:", ruta_json)
```

```python
# Cargar JSON con LangChain: JSONLoader
from langchain_community.document_loaders import JSONLoader

# jq_schema: define qué parte del JSON extraer. Aquí tomamos la lista de clientes.
loader_json = JSONLoader(
    file_path=ruta_json,
    jq_schema=".clientes[]",
    text_content=False
)

docs_json = loader_json.load()
print("Documentos cargados:", len(docs_json))
print("\n--- Vista previa ---\n")
print(docs_json[0].page_content)
```

## Obtener información desde una base de datos SQL (enfoque LangChain)

LangChain usa SQLDatabase.

En LangChain, lo típico para SQL no es un loader en langchain_community.document_loaders, sino la abstracción:

* langchain_community.utilities.SQLDatabase

Aquí creamos una BD, insertamos datos, consultamos con SQL y luego opcionalmente pedimos a Gemini que resuma o analice los resultados.

```python
!pip -q install -U langchain langchain-community sqlalchemy
```

```python
from sqlalchemy import create_engine, text
from langchain_community.utilities import SQLDatabase

# 1) Crear la BD (igual que tú)
engine = create_engine("sqlite:///ejemplo.db")
with engine.begin() as conn:
    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS pedidos (
            id INTEGER PRIMARY KEY,
            cliente TEXT,
            ciudad TEXT,
            total INTEGER
        )
    """))
    conn.execute(text("DELETE FROM pedidos"))
    conn.execute(text("""
        INSERT INTO pedidos (cliente, ciudad, total) VALUES
        ('Ana', 'Medellín', 350),
        ('Luis', 'Bogotá', 120),
        ('Sofía', 'Cali', 280),
        ('Ana', 'Medellín', 410)
    """))

# 2) Conectar vía LangChain
db = SQLDatabase.from_uri("sqlite:///ejemplo.db")

# 3) Ejecutar consulta y obtener resultado (texto)
texto_sql = db.run("SELECT ciudad, COUNT(*) as n, SUM(total) as total FROM pedidos GROUP BY ciudad;")
print(texto_sql)
```

✅ Esto es “LangChain-native” para SQL.

## Mini-ejemplo: usar Gemini para resumir cualquiera de las fuentes

A continuación un ejemplo genérico: tomamos el contenido de DOCX, Excel, JSON y SQL y le pedimos a Gemini un resumen.


```python
# Unificar contenido (DOCX + Excel + JSON + SQL) como contexto
contexto = []
contexto.append("=== DOCX ===\n" + docs_docx[0].page_content)
contexto.append("=== EXCEL ===\n" + texto_excel)
contexto.append("=== JSON ===\n" + "\n".join([d.page_content for d in docs_json]))
contexto.append("=== SQL ===\n" + texto_sql)

contexto_total = "\n\n".join(contexto)

prompt = f"""Eres un asistente académico.
Con el siguiente contexto, escribe un resumen de 6-10 líneas y luego 3 insights accionables.

CONTEXTO:
{contexto_total}
"""

# Requiere que ya esté configurado tu modelo Gemini en las celdas anteriores
respuesta = llm.invoke(prompt)
print(respuesta.content if hasattr(respuesta, "content") else respuesta)
```
