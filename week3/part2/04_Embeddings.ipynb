## 2. Incrustación de texto (embeddings)

```python
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings

embeddings = FastEmbedEmbeddings(model_name="BAAI/bge-small-en-v1.5")

vec = embeddings.embed_query("Esto es un texto para convertirlo en un vector n-dimensional")
print("Dim:", len(vec))
print(vec[:8])
```

## 2. Incrustación de documentos desde CSV

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(
    'datos_ventas_small.csv',
    csv_args={'delimiter': ';'}
)

data = loader.load()

type(data), type(data[0])
```

## 3. Extraer contenido de documentos antes de incrustar

```python
documentos_texto = [d.page_content for d in data]

documentos_texto[:2]  # Mostrar una vista previa
```

## 4. Incrustar documentos con Gemini

```python
embedded_docs = embeddings.embed_documents(documentos_texto)

len(embedded_docs)
```
